# Conditional Probability


---


Conditional probability is the probability of an event A occurring given that another event B has occurred.

image.png

The formula for conditional probability is:


image.png

**Example:**

Two types of cake are sold in a sweet shop: strawberry and chocolate.
By recording the preferences of 50 clients of both genders, the following values ​​were determined:

- 27 women: 11 of whom prefer strawberry cake and 16 chocolate.

- 23 men: 15 choose chocolate and 8 strawberries.


What is the probability of choosing a chocolate cake knowing that the client is a woman?

**Solution:**

Using a contingency table the totals are:

image.png

Then the favorable cases are observed, and first we define the events:

-W is the "woman client" event.

-C is the event "prefer chocolate cake".


Using the conditional probability formula:

P(C│W) = P(C ∩ W) / P(W)

P(W) = 27/50 probability of a woman client.

P(C ∩ W) = 16/50 

Where P(C ∩ W) is the probability that the customer prefers chocolate and is a woman. Now the values ​​are substituted:

P(C│W) = P(C ∩ W) / P(W) = (16/50) / (27/50) = 16/27 = 0.5926.

The probability of a woman choosing a chocolate cake is 59.26%.


```python
import pandas as pd
import numpy as np

#create pandas DataFrame with raw data
df = pd.DataFrame({'Gender': np.repeat(np.array(['Men', 'Women']), (23,27)),
                   'Cake type': np.repeat(np.array(['Chocolate','Strawberry', 'Chocolate','Strawberry']), (15, 8, 16, 11))})

#produce contingency table to summarize raw data
contingency_table = pd.crosstab(index=df['Gender'], columns=df['Cake type'], margins=True)

#view contingency table
contingency_table
```

```python

#calculate probability of choosing a chocolate cake knowing that the client is a woman
p_w=contingency_table.iloc[1, 2]/contingency_table.iloc[2, 2]
print("P(W)=",p_w)
p_c_and_w = contingency_table.iloc[1, 0]/contingency_table.iloc[2, 2]
print("P(C ∩ W)=",p_c_and_w)

p_cw= p_c_and_w/p_w
print("P(C│W)=",round(p_cw,4))
print('The probability of a woman choosing a chocolate cake is %.2f%%' % (p_cw * 100))
```

# **Bayes’ Theorem**


---



Bayes’ theorem describes the probability of occurrence of an event related to any condition.

The formula to calculate bayes theorem is:

where P(B) ≠ 0

**Example:**

A doctor is interested in finding out a patient's probability of **having liver disease if they are an alcoholic**. 

"Being an alcoholic" is the test for liver disease.

 

*   Past data tells the doctor that 10% of patients entering the clinic have liver disease.
*   5% of the clinic's patients are alcoholics. 

*   The doctor also knows that among those patients diagnosed with liver disease, 7% are alcoholics.

**Solution:**

Let

A be the event of “Patient has liver disease."
p(A) = 0.1

B be the event of “Patient is an alcoholic.”
p(B) = 0.05

B|A the probability that a patient is alcoholic, given that they have liver disease.
p(B|A) = 0.07

p(A|B) = [p(B|A)p(A)]/p(B) = [(0.07)(0.1)]/0.05 = 0.14 =14%

In other words, if the patient is an alcoholic, their chances of having liver disease is (14%). 

```python
# calculate the probability of patient having liver disease if they are an alcoholic
 
# calculate P(A|B) given P(A), P(B|A), P(B)
def bayes_theorem(p_a, p_b_given_a, p_b):
	# calculate P(A|B)
	p_a_given_b = (p_b_given_a * p_a) / p_b
	return p_a_given_b

# P(A)
p_a = 0.1
# P(B)
p_b = 0.05
# P(B|A)
p_b_given_a = 0.07

# calculate P(A|B)
p_a_given_b = bayes_theorem(p_a, p_b_given_a, p_b)
print("P(A|B) = %.2f%%" %(p_a_given_b))
print('The probability that an alcoholic patient will have liver disease is = %.f%%' % (p_a_given_b * 100))
```
